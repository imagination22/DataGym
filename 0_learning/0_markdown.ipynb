# ✍️ Markdown Cheat Sheet — For Databricks, DataGym, and Beyond

Markdown is like the minimalist design language of writing—just a few symbols, and boom, your plain text turns into beautifully formatted content. Let’s break it down like a cheat sheet you’d sketch for DataGym.

---

## 🧾 What Is Markdown?

**Markdown** is a lightweight markup language that lets you format text using simple characters like `#`, `*`, and `|`. It’s used in notebooks (like Databricks), GitHub READMEs, blogs, and documentation.

**Normal text** is just raw writing—no formatting unless the platform adds it.  
**Markdown** gives you control: headings, bold, italics, lists, tables, links, and more.

---

## 🧠 Markdown Basics — Think of It Like a Syntax Toolkit

### 🔹 Headings

Use `#` for headings. More `#` means smaller heading.

```markdown
# Heading 1
## Heading 2
### Heading 3
```

### 🔹 Bold and Italics

```markdown
**Bold text**
*Italic text*
```

You can also use `_underscores_` for italics and `__double underscores__` for bold.

### 🔹 Lists

**Unordered list** (bullets):

```markdown
- Item A
- Item B
  - Sub-item B1
```

**Ordered list** (numbered):

```markdown
1. Step one
2. Step two
```

### 🔹 Links

```markdown
[Click here](https://example.com)
```

### 🔹 Tables

Use pipes `|` and dashes `-` to create tables:

```markdown
| Name     | Role         |
|----------|--------------|
| Shubham  | Data Engineer|
| Copilot  | AI Companion |
```

### 🔹 Code Blocks

Use triple backticks for code:

    ```python
    print("Hello, Markdown!")
    ```

Or single backticks for inline code:  
`SELECT * FROM table`

## 🧪 Markdown vs Normal Text — Quick Comparison

| Feature        | Normal Text | Markdown |
|----------------|-------------|----------|
| Headings       | ❌ No        | ✅ Yes (`#`) |
| Bold/Italics   | ❌ No        | ✅ Yes (`**`, `*`) |
| Lists          | ❌ No        | ✅ Yes (`-`, `1.`) |
| Tables         | ❌ No        | ✅ Yes (`|`) |
| Links          | ❌ No        | ✅ Yes (`[text](url)`) |
| Code Formatting| ❌ No        | ✅ Yes (```) |

## 🧰 Where Can You Use Markdown?

- Databricks notebooks (Markdown cells)
- GitHub README files (`.md`)
- Jupyter notebooks
- Documentation tools (e.g., Docusaurus, MkDocs)
- Blogging platforms (e.g., Ghost, Hugo)

---

## 🪄 Pro Tips

- Always leave a blank line before and after headings or lists for clean rendering.
- Use preview mode in Databricks to see how your Markdown looks.
- You can mix Markdown and code in notebooks for beautiful documentation.
